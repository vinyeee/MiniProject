```python
#계절별 그래프 제작 Start

#2017,2018,2019년도 봄

#2017
spring17=weather17.iloc[80:172]#3월22일~6월21일 
spring17=spring17.loc[:,['날짜','평균기온','평균풍속(m/s)','평균습도']]
spring17.reset_index(inplace=True,drop=True)
spring17.columns=['날짜','평균기온17','평균풍속(m/s)17','평균습도17']
spring17['날짜']=[d[:6]for d in spring17['날짜']]
spring17
```

```python
#2018
spring18=weather18.iloc[80:172]#3월22일~6월21일 
spring18=spring18.loc[:,['날짜','평균기온','평균풍속(m/s)','평균습도']]
spring18.reset_index(inplace=True,drop=True)
spring18.columns=['날짜','평균기온18','평균풍속(m/s)18','평균습도18']
spring18['날짜']=[d[:6]for d in spring18['날짜']]
spring18
```

```python
#2019 
spring19=weather19.iloc[80:172]#3월22일~6월21일 
spring19=spring19.loc[:,['날짜','평균기온','평균풍속(m/s)','평균습도']]
spring19.reset_index(inplace=True,drop=True)
spring19.columns=['날짜','평균기온19','평균풍속(m/s)19','평균습도19']
spring19['날짜']=[d[:6]for d in spring19['날짜']]
spring19
```

```python
spring=pd.merge(spring17,spring18)
spring=pd.merge(spring,spring19)
spring
```

```python
#일별 봄 평균 체감온도 
mt=(spring['평균기온17']+spring['평균기온18']+spring['평균기온19'])/3
mw=(spring['평균습도17']+spring['평균습도18']+spring['평균습도19'])/3
spring['평균 체감온도']=round((13.12 + 0.6215 * mt) - (11.37 * mw ** 0.16) + (0.3965 * mt * mw ** 0.16),2)
spring
```

```python
fig_spring=plt.figure(figsize=(60,16))
plt.rc("font", size=20)
plt.plot('날짜','평균 체감온도','k--o',data=spring)
plt.plot('날짜',[17 for i in spring['날짜']],'r',data=spring,label='17℃')
plt.plot('날짜',[23 for i in spring['날짜']],'r',data=spring,label='23℃')
plt.xticks(rotation=50)

plt.title("봄평균 체감온도 (2017~2019 3개년 평균치)")
plt.xlabel("날짜")
plt.ylabel("평균 체감온도")
plt.grid(True)
plt.legend()

plt.show()
```

```python
new_spring=spring.loc[spring['평균 체감온도'].between(17,23)]
new_spring
```

```python
fig_spring=plt.figure(figsize=(40,16))
plt.rc("font", size=18)
plt.plot('날짜','평균 체감온도','r--o',data=new_spring)
plt.xticks(rotation=50)

plt.title("봄평균 체감온도 17℃~23℃ (2017~2019 3개년 평균치)")
plt.xlabel("날짜")
plt.ylabel("평균 체감온도")
plt.grid(True)
plt.legend()

plt.show()
```
