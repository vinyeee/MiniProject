### 데이터 전처리

```python
resort = df[df['hotel']=='Resort Hotel']
resort
```

### 값이 객체 타입인 컬럼만 모아 인코딩 - 리조트 호텔

```python
resort_cat_cols = [col for col in resort.columns if resort[col].dtype == 'O']
resort_cat_cols
```

```python
resort_cat_df = resort[resort_cat_cols]
resort_cat_df.head()
```

```python
resort_cat_df['reservation_status_date'] = pd.to_datetime(resort_cat_df['reservation_status_date'])

resort_cat_df['year'] = resort_cat_df['reservation_status_date'].dt.year
resort_cat_df['month'] = resort_cat_df['reservation_status_date'].dt.month
resort_cat_df['day'] = resort_cat_df['reservation_status_date'].dt.day
```

```python
resort_cat_df.drop(['reservation_status_date','arrival_date_month'] , axis = 1, inplace = True)
resort_cat_df.head()
```

```python
resort_cat_df['hotel'].value_counts()
```

```python
# printing unique values of each column
for col in resort_cat_df.columns:
    print(f"{col}: \n{resort_cat_df[col].unique()}\n")
```

```python
resort_cat_df['hotel'] = resort_cat_df['hotel'].map({'Resort Hotel' : 0})

resort_cat_df['meal'] = resort_cat_df['meal'].map({'BB' : 0, 'FB': 1, 'HB': 2, 'SC': 3, 'Undefined': 4})

resort_cat_df['market_segment'] = resort_cat_df['market_segment'].map({'Direct': 0, 'Corporate': 1, 'Online TA': 2, 'Offline TA/TO': 3,
                                                           'Complementary': 4, 'Groups': 5, 'Undefined': 6, 'Aviation': 7})

resort_cat_df['distribution_channel'] = resort_cat_df['distribution_channel'].map({'Direct': 0, 'Corporate': 1, 'TA/TO': 2, 'Undefined': 3,
                                                                       'GDS': 4})

resort_cat_df['reserved_room_type'] = resort_cat_df['reserved_room_type'].map({'C': 0, 'A': 1, 'D': 2, 'E': 3, 'G': 4, 'F': 5, 'H': 6,
                                                                   'L': 7, 'B': 8})

resort_cat_df['deposit_type'] = resort_cat_df['deposit_type'].map({'No Deposit': 0, 'Refundable': 1, 'Non Refund': 3})

resort_cat_df['customer_type'] = resort_cat_df['customer_type'].map({'Transient': 0, 'Contract': 1, 'Transient-Party': 2, 'Group': 3})

resort_cat_df['year'] = resort_cat_df['year'].map({2015: 0, 2014: 1, 2016: 2, 2017: 3})
```

```python
resort_cat_df.head()
```

### 값이 숫자 타입인 컬럼만 모아 데이터프레임 정제 - 리조트 호텔

```python
resort_num_df = resort.drop(columns = resort_cat_cols, axis = 1)
resort_num_df.drop('is_canceled', axis = 1, inplace = True)
resort_num_df
```

```python
resort_num_df.var()
```

### 표준화 

* 분산 값이 큰 피쳐들을 대상으로 log를 취해 nomalizing 해준다.


* Log를 취해주게 되면 큰 숫자를 같은 비율의 작은 숫자로 만들어주며 정규성이 높아진다.이는 분석의 정확도를 더 높일 수 있는 결과를 가져온다. 

```python
# normalizing numerical variables

resort_num_df['lead_time'] = np.log(resort_num_df['lead_time'] + 1)
resort_num_df['arrival_date_week_number'] = np.log(resort_num_df['arrival_date_week_number'] + 1)
resort_num_df['arrival_date_day_of_month'] = np.log(resort_num_df['arrival_date_day_of_month'] + 1)
resort_num_df['agent'] = np.log(resort_num_df['agent'] + 1)
resort_num_df['company'] = np.log(resort_num_df['company'] + 1)
resort_num_df['adr'] = np.log(resort_num_df['adr'] + 1)
```

```python
resort_num_df.var()
```

```python
resort_num_df.isnull().sum()
```

```python
resort_num_df['adr']=resort_num_df['adr'].fillna(value=resort_num_df['adr'].mean())
resort_num_df.head()
```

### 객체 타입 데이터와 숫자 타입 데이터를 합친다. - 리조트 호텔

```python
X = pd.concat([resort_cat_df,resort_num_df],axis = 1)
y = resort['is_canceled']
```

```python
X.shape,y.shape
```

### 리조트 호텔 상관계수

```python
def cal_corr(df):
    ### # 그림 사이즈 지정
    fig= plt.subplots( figsize=(24, 20) )


    # 삼각형 마스크를 만든다(위 쪽 삼각형에 True, 아래 삼각형에 False)
    mask = np.zeros_like(df.corr(), dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True
    
    # 히트맵을 그린다
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    sns.heatmap(df.corr(),
                cmap = colormap,
                annot = True,   # 실제 값을 표시한다
                mask=mask,      # 표시하지 않을 마스크 부분을 지정한다
                linewidths=.5,  # 경계면 실선으로 구분하기
                cbar_kws={"shrink": .5},# 컬러바 크기 절반으로 줄이기
                vmin = -1,vmax = 1)   # 컬러바 범위 -1 ~ 1  
   
    
cal_corr(pd.concat([X,y],axis = 1))

```

### 데이터셋 분리

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

### 모델생성

###  GBM -리조트 호텔

```python
from sklearn.ensemble import GradientBoostingClassifier
gb= GradientBoostingClassifier()
gb.fit(X_train,y_train) #순차처리이기 때문에 n_jobs가 지원안됨
```

### 예측

```python
from sklearn.metrics import accuracy_score, confusion_matrix,f1_score
```

```python
pred_gb = gb.predict(X_test)
pred_gb
```

### 평가

```python
gb.score(X_train,y_train)
```

```python
gb.score(X_test,y_test)
```

```python
def result(y_test,pred) :
    acc_lr = accuracy_score(y_test,pred)
    conf = confusion_matrix(y_test,pred)
    f1score = f1_score(y_test,pred)
    print(f'정확도: {acc_lr}')
    print(f"Confusion Matrix : \n{conf}")
    print(f'f1 score :{f1score} ')

result(y_test,pred_gb)    
```
