### 1. 나라별 예약 취소 분석

```python
resort_guest = df[df['hotel'] == 'Resort Hotel']
resort_guest = resort_guest.loc[:,['is_canceled','country']]
resort_guest = resort_guest[resort_guest['is_canceled'] == 1]['country'].value_counts().reset_index()
resort_guest.columns = ['country','canceled']
resort_guest['hotel']=['Resort Hotel' for i in resort_guest['country']] 
resort_guest
```

```python
city_guest = df[df['hotel'] == 'City Hotel']
city_guest = city_guest.loc[:,['is_canceled','country']]
city_guest = city_guest[city_guest['is_canceled'] == 1]['country'].value_counts().reset_index()
city_guest.columns = ['country','canceled']
city_guest['hotel']=['City Hotel' for i in city_guest['country']] 
city_guest
```

```python
country_guest = pd.concat([resort_guest,city_guest],ignore_index=True)
country_guest
```

```python
#!pip install plotly
import folium
import plotly.express as px

px.bar(data_frame = country_guest, x = 'country', y = 'canceled', barmode = 'group', color= 'hotel',template = 'plotly_dark')
```

### 3. 1년 동안 1박당 가격의 변화

```python
data_resort = data[data['hotel'] == 'Resort Hotel']
data_city = data[data['hotel'] == 'City Hotel']
data_resort
```

```python
resort_hotel = data_resort.groupby(['arrival_date_month'])
resort_hotel = resort_hotel['adr'].mean().reset_index()
resort_hotel
```

```python
city_hotel = data_city.groupby(['arrival_date_month'])
city_hotel = city_hotel['adr'].mean().reset_index()
city_hotel
```

```python
final_hotel = resort_hotel.merge(city_hotel,on='arrival_date_month')
final_hotel.columns=['month','price_for_resort_hotel','price_for_city_hotel']
final_hotel
```

### month 열의 월이 순서대로 정렬되어있지 않다. => 순서대로 정렬


```python
#!pip install sort-dataframeby-monthorweek

#!pip install sorted-months-weekdays

import sort_dataframeby_monthorweek as sd
    
def sort_month(df, column_name):
    return sd.Sort_Dataframeby_Month(df, column_name)

final_prices = sort_month(final_hotel,'month')
final_prices
```

```python
plt.figure(figsize = (17,8))
px.line(final_prices, x = 'month',
        y= ['price_for_resort_hotel','price_for_city_hotel'],
       title='Room price per night over the Months',
       template='plotly_dark')
```

### 리조트 호텔의 가격이 여름에 훨씬 더 높고 시티 호텔의 가격은 비교적 평이하며 봄과 가을에 가장 비싸다.

### 4. 가장 바쁜 달은 언제인가?

```python
resort_guests = data_resort['arrival_date_month'].value_counts().reset_index()
resort_guests.columns=['month','no of guests']
resort_guests
```

```python
city_guests = data_city['arrival_date_month'].value_counts().reset_index()
city_guests.columns=['month','no of guests']
city_guests
```

```python
final_guests = resort_guests.merge(city_guests,on='month')
final_guests.columns=['month','no of guests in resort','no of guest in city hotel']
final_guests
```

```python
final_guests = sort_month(final_guests,'month')
final_guests
```

```python
px.line(final_guests, x='month', y=['no of guests in resort','no of guest in city hotel'],
        title='Total no of guests per Months',template='plotly_dark')
```
